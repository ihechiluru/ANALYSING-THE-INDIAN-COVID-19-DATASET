# **Does Lockdown Really Help in curtailing the virus?**

```python
india_df['month'] = pd.DatetimeIndex(india_df.Date).month
```

```python
india_df.head()
```

```python
india_month_df = india_df.groupby('month')[['Active_cases', 'Confirmed', 'Cured', 'Deaths',]].sum()
```

```python
india_month_df
```

Now even though this gives us an awesome amount of insight into the intensity of the metrics per month, we have to keep in mind that our data has two years in it (2020 and 2021).


Therefore, the values we are getting are the sum of values over the same month in both years 

```python
india_year_df = india_df.groupby('year', as_index=True)[['Active_cases', 'Confirmed', 'Cured', 'Deaths','States']].sum()
```

```python
india_year_df
```

```python
sns.lineplot( data = india_year_df, x = 'year', y = 'Active_cases');
```

Looking at the graph plot above we can see that the Active cases in india increased over time from 2020 to 2020. 

Now that we have this information, we would also like to know how the trend of active cases was in 2021 and at which points it was highest as well as lowest.

This analysis is given us an ample amount of insight of the active cases in out data.

Now, we know that The active cases were relatively lower from January till somewhere around March, having Febuary as the month with the lowest point.

Then, there is a very fast pick up in the number of Active cases in May and this upward trend in Active cases peaks in the month of May with average cases at over 100000.

Thereafter, there is a decrease in number of cases and this downward trend has been on since June.

Next we'll make the same plot for tests around the same time frame (2021) and try to understand what the relation between tests and active cases was like.


```python
statewise_tests_df
```

```python
statewise_tests_df['Date'] = pd.to_datetime(statewise_tests_df.Date)
```

```python
statewise_tests_df['year'] = pd.DatetimeIndex(statewise_tests_df.Date).year
statewise_tests_df['weekday'] = pd.DatetimeIndex(statewise_tests_df.Date).weekday
statewise_tests_df['day'] = pd.DatetimeIndex(statewise_tests_df.Date).day
```

```python
tests_2021 = statewise_tests_df[statewise_tests_df.year == 2021]
tests_2021
```

```python
tests_2021_metrics = tests_2021[['Date','TotalSamples', 'Negative', 'Positive']]
tests_2021_metrics
```

```python
fig = plt.figure(figsize=(14,7))
plt.title('Trend for Confirmed Tests in 2021')
sns.lineplot (data= tests_2021_metrics, x = 'Date', y = 'TotalSamples');
```

Looking at this plot and especially in retrospect to the plot we've made earlier analysing the Active cases, we would notice that the relationship between the tests conducted and the Active cases are not linear. 

The tests over time only increased whilst the active cases on the other hand had a curve (increased and at some point started reducing). 


---



Now, let us seek insight into the relationship between the confirmed tests and positve results.

```python
with plt.xkcd():
  fig = plt.figure(figsize=(14,7))
  plt.title('Positive Results From Statewise Tests')
  sns.lineplot (data= tests_2021_metrics, x = 'Date', y = 'Positive');
```

This shows that even though there was an upward trend in the plot for Total Tests, The tests were not directly proportional to the Positive results.

Even looking at the plots, Febuary (the month with the lowest total tests) and August the month with the highest has similar amounts of positive cases.
